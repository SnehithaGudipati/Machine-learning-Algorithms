# **SVM**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df1=pd.read_csv("/content/Social_Network_Ads.csv")
df1
```

```python
x= df1.iloc [:,2:4].values
print (x)
```

```python
y = df1.iloc[:,4].values
print (y)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x= StandardScaler()
x = sc_x.fit_transform(x)
print (x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='sigmoid')
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x)
y_pred
```

```python
df_x=pd.DataFrame(x,columns=['Age','Salary'])
df_y=pd.DataFrame(y,columns=['Y'])
df_ypred=pd.DataFrame(y_pred,columns=['Y Pred'])
result=pd.concat([df_x,df_y,df_ypred],axis=1)
print(result)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y,y_pred)
print(acc)
```

```python
classifier=SVC(kernel='poly')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x)
print(accuracy_score(y,y_pred))
#print(classifcation_report(y,y_pred))
cm=confusion_matrix(y,y_pred)
print(cm)
```

```python
classifier=SVC(kernel='linear')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x)
print(accuracy_score(y,y_pred))
#print(classifcation_report(y,y_pred))
cm=confusion_matrix(y,y_pred)
print(cm)
```
