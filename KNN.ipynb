# **KNN Classification**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df1=pd.read_csv("/content/Social_Network_Ads.csv")
df1
```

```python
x=df1.iloc[:,2:4].values
y=df1.iloc[:,-1].values
print(x)
print(y)
```

```python
plt.scatter(x[y==0,0],x[y==0,1],s=100,c='g',label='cluster 1')
plt.scatter(x[y==1,0],x[y==1,1],s=100,c='r',label='cluster 2')
plt.show()
```

```python
df_try=pd.DataFrame({"Age":[23,25,24,26],"Salary":[23000,24000,25000,26000],"Purchase":[1,0,1,0]})
```

```python
x1=df_try.iloc[:,0:2].values
y1=df_try.iloc[:,2].values
print(x1)
print(y1)
```

```python
plt.scatter(x1[y1==0,0],x1[y1==0,1],s=100,c='g',label='cluster 1')
plt.scatter(x1[y1==1,0],x1[y1==1,1],s=100,c='r',label='cluster 2')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x = sc_x.fit_transform(x)
print (x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
```

```python
y_pred=KNN.predict(x_test)
print(y_pred)
```

```python
y_pred_f=KNN.predict(x)
y_pred_f
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score (y,y_pred_f)
print (accuracy)
```

```python
cm=confusion_matrix(y,y_pred_f)
print(cm)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN1 = KNeighborsClassifier(n_neighbors =1 , metric = 'minkowski', p= 1)
KNN1.fit(x_train,y_train)
print (KNN1)
```

```python
y_pred=KNN1.predict(x)
print(y_pred)
```

```python
accuracy=accuracy_score(y,y_pred)
print(accuracy)
```
